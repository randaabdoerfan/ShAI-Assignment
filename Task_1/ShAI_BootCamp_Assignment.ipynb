# 3-Data Cleaning

```python
#3-after show the dataset we notice that all the values in columns {Notes and Status} have NAN values so we should drop them from the data set to clean it 
to_drop=["Notes","Status"]

df.drop(to_drop ,inplace=True , axis =1)

df.head()
```

```python
df.columns
```

```python
# we noticed that there is colums {OtherPays ,Benefits ,BasePay} have some NAN values so to clean the data prefect we should put values for these NAN values like mean of them 
cols =["BasePay" ,"OvertimePay" ,"OtherPay" ,"Benefits"]

for col in cols :
    df[col] = df[col].fillna(df[col].mean())
    
#to make sure that all columns don't have any NAN values
df.isnull().sum()
```

# 4-Basic Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
#using basic histogram to show the distrbution of salaries
plt.hist(df.TotalPay,bins=50 ,edgecolor="black" ,color="#83f442")

plt.title("Distribution of Salaries")

plt.xlabel("Salaries")

plt.ylabel("number of people")

plt.show()
```

```python
# to see hw many different job in the dataset 
len(df['JobTitle'].unique())
```

```python
grouped = df.groupby('JobTitle').count()
top5 = grouped.sort_values(by='Id',  ascending=False)[:20]
top5['Id']
```

```python
#cause the jobs are alot as we see above so we will take the most 20 popular of them to make the pie 
plt.figure(figsize=(10,10))

df["JobTitle"].value_counts()[:20].plot(kind="pie", autopct='%.2f',title="Job Title",label =" ")

plt.legend(loc=(0.9,0.9))

plt.show()
```

# 5-Group Analysis

```python
#group of jobs acording to the total pay
df[["TotalPay","JobTitle"]].groupby("JobTitle").mean()
```

```python
#group by job accorrding to totalpaybenefits
df[["TotalPayBenefits","JobTitle"]].groupby("JobTitle").mean()
```

```python
#Group by two keys and then summarize each group
dept_gender_salary =  df.groupby(['EmployeeName','JobTitle'],as_index=False).TotalPay.mean()
print(dept_gender_salary)
```

```python
plt.figure(figsize=(15,8))
plt.title("number of people in each department")
jobs= df.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending =False)[:50]
x = jobs.index
plt.bar(x,jobs, width=0.6)
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
plt.title("number of people in each department")
jobs= df.groupby('JobTitle')['TotalPayBenefits'].mean().sort_values(ascending =False)[:50]
x = jobs.index
plt.bar(x,jobs, width=0.6)
plt.xticks(rotation=90)
plt.show()
```
